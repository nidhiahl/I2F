```python
import pandas as pd
from sklearn.datasets import make_moons, make_blobs
import numpy as np
import matplotlib as mpl

from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

```python
# Example settings
n_samples = 500
outliers_fraction = 0.015
n_outliers = int(outliers_fraction * n_samples)
n_inliners = n_samples - n_outliers
```

```python
blobs_params = dict(random_state=0, n_samples=n_inliners, n_features=2)
```

```python
blob_ds1= make_blobs(centers=[[2, 2],[3,3], ], cluster_std=[0.25,0.25], **blobs_params)[0]
```

```python
anomalyclass = np.array([0] * n_inliners)
```

```python
len(anomalyclass)
```

```python
ds1=np.column_stack((blob_ds1,anomalyclass))
```

```python
anomalies = np.random.uniform(low=0, high=3.5, size=(int(1*n_outliers), 2))

anomalyclass = np.array([1] * int(1*n_outliers))

anomalies=np.column_stack((anomalies,anomalyclass))

# Add outliers
ds1 = np.concatenate([ds1,anomalies],axis = 0)
```

```python
data = pd.DataFrame(ds1)
```

```python
x=data[0]
y=data[1]
clusters = data[2]
```

```python
cmap = plt.cm.jet  # define the colormap
# extract all colors from the .jet map
cmaplist = [cmap(i) for i in range(cmap.N)]
# force the first color entry to be grey
cmaplist[0] = (.5, .5, .5, 1.0)
```

```python
# create the new map
cmap = mpl.colors.LinearSegmentedColormap.from_list(
    'Custom cmap', cmaplist, cmap.N)
```

```python
plt.scatter(x,y,c=clusters,s=15,cmap=cmap)

cbar=plt.colorbar()



plt.tight_layout()

plt.show()
```

```python
anomalyclass = np.array([2] * 2)
```
