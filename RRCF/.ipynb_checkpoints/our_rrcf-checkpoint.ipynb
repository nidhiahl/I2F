```python
import pandas as pd
import numpy as np
import rrcf
def load_data(dataFile: str):
  df = pd.read_csv(dataFile,sep=' ')
  df = df.astype(float)
  class_labels = df.iloc[:, -1:]
  class_labels = class_labels.to_numpy()
  df = df.iloc[: , :-1]
  df = df.to_numpy()
  return df, class_labels


def load_data_without_labels(dataFile: str):
  df = pd.read_csv(dataFile,sep=' ')
  df = df.astype(float)
  df = df.to_numpy()
  return df
```

```python

data = pd.read_csv('/content/gdrive/My Drive/nyc_taxi_data/nyc_taxi_shingled.csv', skiprows=1,sep=' ',header=None)
```

```python
data=data.astype(float)
data=data.to_numpy()
D0=data[:10000]
```

```python
# Set forest parameters
num_trees = 100
tree_size = 256
nD0 = len(D0)
sample_size_range = (nD0 // tree_size, tree_size)

# Construct forest
forest = []
while len(forest) < num_trees:
    # Select random subsets of points uniformly
    ixs = np.random.choice(nD0, size=sample_size_range,
                           replace=False)
    # Add sampled trees to forest
    trees = [rrcf.RCTree(D0[ix], index_labels=ix)
             for ix in ixs]
    forest.extend(trees)
```

```python
deltaD=data[10000:]
ndeltaD=len(deltaD)
avg_codisp = {}
avg_codisp=pd.Series(0.0, index=np.arange(ndeltaD))
avg_codisp
for index, point in enumerate(data[10000:]):
    # For each tree in the forest...
    for tree in forest:
        print(index,point)
        # If tree is above permitted size...
        #if len(tree.leaves) > tree_size:
            # Drop the oldest point (FIFO)
            #tree.forget_point(index+10000 - tree_size)
        # Insert the new point into the tree
        tree.insert_point(point, index=index+10000)
        # Compute codisp on the new point...
        new_codisp = tree.codisp(index+10000)
        # And take the average over all trees
        if not index in avg_codisp:
            avg_codisp[index] = 0
        avg_codisp[index] += new_codisp / num_trees
```
