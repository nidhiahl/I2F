```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset='synthetic'
update_type='CDIII'
testset_name='Di'

#methods=['lazy','i2f','aggressive','iFADS','PCBIF','SENCForest']

methods=['aggressive','i2f','iFADS','PCBIF','SENCForest','RRCF']

df = pd.DataFrame()
df
for i, method in enumerate(methods):
    df[method] = pd.read_csv('results/'+dataset+'/'+update_type+'/'+method+'_avg_ascore_'+testset_name+'.csv',sep=' ',header=None)

if testset_name == 'Di':
    file=''
else:
    file='_'+update_type
    
df_data = pd.read_csv('datasets/'+dataset+'/'+testset_name+file+'.csv',sep=' ',header=None,skiprows=1)

# Create subplots for each column
#fig, axs = plt.subplots(1, len(methods), figsize=(24, 3), sharey=True)
#plot_tilte = ['lazy','i2f','aggressive','iFASD','PCBIF','SENCForest']
plot_tilte = ['aggressive \n[Ideal Results]','i2f','iFASD','PCBIF','SENCForest','RRCF']
colors = ['blue','green','red']

# Plot each column

fig = plt.figure(figsize=(24, 3))
ax1 = fig.add_subplot(161)
mesh1 = ax1.scatter(df_data[0], df_data[1], c=df[methods[0]], cmap='rainbow',s=5)
ax1.set_title(plot_tilte[0])
fig.colorbar(mesh1,ax=ax1)
ax2 = fig.add_subplot(1,6,2, sharex=ax1)
mesh2 = ax2.scatter(df_data[0], df_data[1], c=df[methods[1]], cmap='rainbow',s=5)
ax2.set_title(plot_tilte[1])
ax2.yaxis.set_visible(False) 
fig.colorbar(mesh2,ax=ax2)
ax3 = fig.add_subplot(163)
mesh3 = ax3.scatter(df_data[0], df_data[1], c=df[methods[2]], cmap='rainbow',s=5)
ax3.set_title(plot_tilte[2])
ax3.yaxis.set_visible(False) 
fig.colorbar(mesh3,ax=ax3)
ax4 = fig.add_subplot(164)
mesh4 = ax4.scatter(df_data[0], df_data[1], c=df[methods[3]], cmap='rainbow',s=5)
ax4.set_title(plot_tilte[3])
ax4.yaxis.set_visible(False) 
fig.colorbar(mesh4,ax=ax4)
ax5 = fig.add_subplot(165)
mesh5 = ax5.scatter(df_data[0], df_data[1], c=df[methods[4]], cmap='rainbow',s=5)
ax5.set_title(plot_tilte[4])
ax5.yaxis.set_visible(False) 
fig.colorbar(mesh5,ax=ax5)
ax6 = fig.add_subplot(166)
mesh6 = ax6.scatter(df_data[0], df_data[1], c=df[methods[5]], cmap='rainbow',s=5)
ax6.set_title(plot_tilte[5])
ax6.yaxis.set_visible(False) 
fig.colorbar(mesh6,ax=ax6,label='Anomlay Score')

plt.savefig('figures/'+dataset+'/scatter_'+testset_name+'_'+update_type+'.png',format='png')

plt.show()


```
