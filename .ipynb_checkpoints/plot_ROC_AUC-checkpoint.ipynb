```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics

dataset='synthetic_2D'
update_type='CDIII'
testset_name='Di'
#methods = ['aggressive','i2f','iFADS','PCBIF','SENCForest','RRCF']
methods = ['lazy','aggressive','i2f']

if testset_name == 'Di':
    file=''
else:
    file='_'+update_type
  
df = pd.DataFrame()
label=pd.DataFrame()
for i, method in enumerate(methods):
    #df[method] = pd.read_csv('results/'+dataset+'/'+update_type+'/'+method+'_avg_ascore_'+testset_name+'.csv',sep=' ',header=None)
    #df[method] = pd.read_csv('results/'+dataset+'/'+update_type+'/'+method+'_avg_ascore_Di.csv',sep=' ',header=None)
    #df[method] = pd.concat([df[method],pd.read_csv('results/'+dataset+'/'+update_type+'/'+method+'_avg_ascore_deltaD.csv',sep=' ',header=None)],ignore_index=True)
    
    df[method] = pd.read_csv('io_directory/'+update_type+'/'+method+'_sumScores_Di.csv',sep=' ',header=None)
    df[method] = pd.concat([df[method],pd.read_csv('io_directory/'+update_type+'/'+method+'__sumScores_deltaD.csv',sep=' ',header=None)],ignore_index=True)
    
    label[method] = pd.read_csv('datasets/'+dataset+'/Di_'+update_type+'_labels.csv',sep=' ',header=None)
    label[method] = pd.concat([label[method],pd.read_csv('datasets/'+dataset+'/deltaD_'+update_type+'_labels.csv',sep=' ',header=None)],ignore_index=True)
    

    
plt.figure(figsize=(4, 4))  # Adjust figure size if needed
#fig, axs = plt.subplots(1, len(methods), figsize=(18, 3), sharey=True)
#plot_tilte = ['lazy','i2f','aggressive','iFASD','PCBIF','SENCForest']
#plot_tilte = ['i2f','aggressive','iFASD','PCBIF','SENCForest','RRCF']

for i, method in enumerate(methods):
    fpr, tpr, thresholds = metrics.roc_curve(label[method], df[method], pos_label=1)
    roc_auc = metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{method}(AUC = {roc_auc:.2f})')
    #display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name='example estimator')
    #display.plot()
    #plt.show()
    
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(dataset+'_'+update_type)
plt.legend(loc="lower right")
#plt.savefig('figures/'+dataset+'/ROC_'+testset_name+'_'+update_type+'.png',format='png')
plt.show()
```
